```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("telco_corr_high.csv")
X = data["age"]
Y = data["tenure"]

print(Y.corr(X))

X = data[["age"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


plt.scatter(X_train, y_train)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r_squared}")

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot regression line
plt.plot(X, model.predict(X), color='red')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Model")
plt.show()
```

```python
from sklearn.svm import SVR

# Initialize and train SVR model
model = SVR(kernel='rbf')  # You can use linear, polynomial, or RBF kernels
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r_squared}")

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot regression line
plt.scatter(X, model.predict(X), color='red')

plt.xlabel("X")
plt.ylabel("y")
plt.title("svm Model")
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Transforming the features into polynomial features
poly = PolynomialFeatures(degree=2)  # For quadratic (degree=2)
X_poly = poly.fit_transform(X_train)

# Train the model
model = LinearRegression()
model.fit(X_poly, y_train)

# Make predictions
y_pred = model.predict(poly.transform(X_test))
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(PolynomialFeatures(degree=2).fit_transform(X_test), y_test)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r_squared}")

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot regression line
plt.scatter(X, model.predict(PolynomialFeatures(degree=2).fit_transform(X)), color='red')

plt.xlabel("X")
plt.ylabel("y")
plt.title("poly regression Model")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train decision tree model
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r_squared}")

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot regression line
plt.scatter(X, model.predict(X), color='red')

plt.xlabel("X")
plt.ylabel("y")
plt.title("decision tree Model")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train random forest model
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r_squared = model.score(X_test, y_test)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r_squared}")

# Scatter plot of the data
plt.scatter(X, Y, color='blue')

# Plot regression line
plt.scatter(X, model.predict(X), color='red')

plt.xlabel("X")
plt.ylabel("y")
plt.title("random forest Model")
plt.show()
```
